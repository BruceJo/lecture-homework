## 1. 최소 정렬 횟수(빈칸 채우기)

중복된 숫자가 없는 리스트를 오름차순으로 정렬할 때, 정렬 횟수를 최소화 하고자 합니다.<br>
[5, 2, 3, 4, 1]은 5와 1을 한번만 바꾸면 됩니다.


<hr>

### 매개변수 설명
다음 조건을 만족하는 input_n이 solution 함수의 매개변수로 주어집니다.<br>

  * input_n의 길이는 1이상 9,999 이하입니다.<br>
  * input_n의 원소는 1이상 9,999 범위로 이루어진 숫자형 입니다.<br>
  * input_n의 원소는 중복값이 들어있지 않습니다.<br>

<hr>

### return 값 설명
리스트가 주어졌을때, 오름차순의 최소 정렬 횟수를 구할 수 있도록 주어진 코드의 빈칸을 적절히 채워 전체 코드를 완성해주세요.

<hr>

### 예시
input_1    [5, 2, 3, 4, 1]<br>
output_1 = 1

input_2    [4, 3, 2, 1]<br>
output_2 = 2

```python
# import cannot be used

def solution(arr):
    temp = [0] * (len(arr)+1)
    answer = 0
    
    for i, value in enumerate(arr):
        temp[value] = i
    
    for i in range(len(arr)):
        
        if arr[i] != i+1 :    # if arr[i] != (@@@):
            index = temp[i+1]    # index = temp[@@@]
            arr[i], arr[index] = arr[index], arr[i]
            temp[arr[index]] = index
            
            answer += 1
            
    return answer

# Below is the code to output the test cases.
input_1 = [5, 2, 3, 4, 1]
input_2 = [4, 3, 2, 1]

print('output_1 =', solution(input_1))    # 1
print('output_2 =', solution(input_2))    # 2
```

## 2. 도서관 책반납 시스템 (한줄 바꾸기)
도서관의 운영을 위해 1명에게 한번에 최대 5권을 빌려주도록 설정된 시스템을 만들었습니다. 이 도서관에서 사용자들에게 대여를 해주었던 책의 총 수를 구하려고 합니다.

배열이 다음과 같이 주어진다.<br>
status = [[1,0,7], [1,1,3], [1,0,2]]<br>
[1,0,7] 중 1은 빌린 사람의 id<br>
&emsp;&emsp;&emsp;&emsp;&nbsp;0은 상태 (0: 대여, 1: 반납)<br>
&emsp;&emsp;&emsp;&emsp;&nbsp;7은 빌리거나 반납하는 도서 수<br>

max_limit = 5<br>

위의 매개변수가 주어졌을때 반납을 제외하고 여태까지 총 몇권을 빌려주었는지 알려한다.<br>
  * 1번고객은 7권을 빌리려했지만 최대 5권만 빌려줄수 있었고,<br>
  * 1번고객은 이후 책 3권을 반납하여 1번고객은 2권만 갖고있는 상태이다.<br>
  * 1번고객은 다시 책 2권을 빌렸으므로 총 7권의 책을 도서관이 대여해주었다.<br>

이때, 주어진 코드에서 **한 줄**만 변경해서 모든 입력에 대해 올바르게 동작하도록 수정하세요.<br>

<hr>

### 매개변수 설명
자연수로 이루어진 2중 리스트가 solution 함수의 매개변수로 주어집니다.

  * status는 숫자로만 이루어진 2중리스트입니다.
  * status의 길이는 1부터 9999까지의 범위로 주어질 수 있습니다.
  * 첫번째 원소는 0부터 999까지의 범위로 이루어진 숫자형 있습니다.
  * 두번째 원소는 0: 대여, 1: 반납 으로만 이루어져 있습니다.
  * 세번째 원소는 0부터 9999까지의 범위로 이루어진 숫자형 입니다.
  * max_limit는 1부터 10 이하의 자연수 입니다.

<hr>

### return 값 설명
solution 함수는 주어진 status에 대해 도서관에서 대여해주었던 총 몇권을 대여해줬는지를 return합니다.

<hr>

### 예시
status_1 = [[1,0,7], [1,1,3], [1,0,2]]  
output_1 = 7

status_2 = [[1,0,5], [1,1,1], [2,0,5], [1,0,3], [2,1,4], [1,0,3]]  
output_2 = 11

status_3 = [[1,0,3], [1,0,3], [1,1,1], [1,1,3], [1,0,5]]  
output_3 = 9

```python
# import cannot be used

def solution(status):
    max_limit = 5
    booker = [0 for x in range(0, 10000)]
    answer = 0

    for [_id, _status, _number] in status:

        if _status == 1:
            booker[_id] -= _number
            continue
        rental = min(_number, max_limit - booker[_id])    # rental = min(max_limit, _number - booker[_id])
        answer += rental
        booker[_id] += rental
    return answer

# Below is the code to output the test cases.
status_1 = [[1,0,7], [1,1,3], [1,0,2]]
status_2 = [[1,0,5], [1,1,1], [2,0,5], [1,0,3], [2,1,4], [1,0,3]]
status_3 = [[1,0,3], [1,0,3], [1,1,1], [1,1,3], [1,0,5]]

print('output_1 = ', solution(status_1))    # 7
print('output_2 = ', solution(status_2))    # 11
print('output_3 = ', solution(status_3))    # 9
```

## 3. 캐비넷 정하기(함수 작성)

학생 5명과 2명씩 쓸수있는 캐비넷이 있습니다. 주어진 학생번호을 정렬하여 첫번째 캐비넷부터 두명씩 배정하려고 합니다. 이때, 각 캐비넷에 할당된 학생번호를 더하여 출력해주세요.<br>

input : <br>
&emsp;cabinet = 2,<br>
&emsp;students = [2001, 2005, 2002, 2004, 2003]<br>
output :<br>
&emsp;[4003, 4007, 2005]<br>
<br>
explanation : [[2001 + 2002], [2003 + 2004], [2005]]

<hr>

### 매개변수 설명
캐비넷을 의미하는 cabinet과 학생번호가 담겨있는 students이 solution 함수의 매개변수로 주어집니다.

  * cabinet은 길이는 1부터 10까지의 범위의 자연수로 주어질 수 있습니다.
  * students의 원소는 학생번호를 의미하며, 1000부터 9999까지의 범위로 이루어진 숫자형 입니다.

<hr>

### return 값 설명
solution 함수는 주어진 cabinet과 students에 대해 각 캐비넷에 할당된 학생번호를 더하여 return합니다.

<hr>

### 예시
| cabinet | students | result | explanation |
|---|---|---|---|
| 2 | [2001, 2005, 2002, 2004, 2003] | [4003, 4007, 2005] | [[2001 + 2002], [2003 + 2004], [2005]] |
| 3 | [2507, 2003, 2005, 2010, 2123, 2019, 2018, 2013] | [6018, 6050, 4630] | [[2003 + 2005 + 2010], [2013 + 2018 + 2019], [2123 + 2507]] |
| 2 | [2007, 2003, 2005, 2009, 2008, 2001] | [4004, 4012, 4017] | [[2001 + 2003], [2005 + 2007], [2008 + 2009]] |


```python
# import cannot be used

def solution(students, cabinet):
    answer = []
    re_list = []
    students.sort()
    
    point = len(students)/cabinet
    point = int(point) if point % 1 == 0 else int(point) + 1
    
    for i in range(point):
        re_list.append(students[i*cabinet : i*cabinet+cabinet])
    for i in range(len(re_list)):
        answer.append(sum(re_list[i]))
    return answer

# Below is the code to output the test cases.
students_1 = [2001, 2005, 2002, 2004, 2003]
cabinet_1 = 2

students_2 = [2507, 2003, 2005, 2010, 2123, 2019, 2018, 2013]
cabinet_2 = 3

students_3 = [2007, 2003, 2005, 2009, 2008, 2001]
cabinet_3 = 2

print('output_1 = ', solution(students_1, cabinet_1))
print('output_2 = ', solution(students_2, cabinet_2))
print('output_3 = ', solution(students_3, cabinet_3))

cabinet_1 = 3
students_1 = [2501, 2502, 2503, 2504, 2505, 2506, 2507, 2508]
print('output_1 = ', solution(students_1, cabinet_1))  
```

## 4. 가성비 프린터 (함수 작성)

수중에 갖고있는 금액을 한도로 프린터를 가격대비 가장 좋은것을 사려고 합니다. 지수(가격대비성능)은 `가격/성능`으로 계산합니다. <br><br>
**지수 : 가격/성능<br>**

<hr>

### 매개변수 설명
  * mymony : 수중의 금액<br>
  * printers : [[프린터 성능, 프린터 가격], [프린터 성능, 프린터 가격], ...]<br>

<hr>

### return 값 설명
solution 함수는 주어진 mymony와 printers에 대해 수중의 금액중에서 살 수 있는 가장 좋은 프린터를 찾아 return합니다.
  * 만약, 수중의 금액이 부족하여 살 수 있는 프린터가 하나도 없으면 0을 출력하고<br>
  * 살수 있고, 동일한 지수의 프린터가 여러개라면 그중 첫번째 자리를 return해야합니다.<br>
  * 살수 있고, 가장 좋은 프린터가 있다면 그 자리를 return해야 한다. <br>

<hr>

### 예시
mymoney : 2<br>
printers = [[3,9], [2,6], [1,3]]<br>
위 예시에서 프린터는 총 3대이며, 지수(가격대비성능)은 모두 3으로 동일하지만
첫번째 프린터는 수중의 돈보다 크기때문에 살수없고 2,3번째 프린터를 살 수 있습니다.<br>
살수 있고, 동일한 지수의 프린터가 2개이므로 그중 첫번째 자리인 2를 리턴해야합니다.<br>


```python
# import cannot be used

def solution(printers, mymoney):
    factor = []
    for performance, price in printers:
        if price <= mymoney:
            factor.append(price / performance)
        else:
            factor.append(0)
    
    answer = 0 if sum(factor)==0 else factor.index(max(factor))+1
    
    return answer

# Below is the code to output the test cases.
print('output_1 = ', solution([[3,9], [2,6], [1,3]], 6)) # 2
print('output_2 = ', solution([[3,9], [2,6], [3,3]], 2)) # 0
print('output_3 = ', solution([[3,9], [4,6], [1,3]], 4)) # 3
print('output_4 = ', solution([[2, 5], [4, 10], [4, 7]], 10))  # 1
```

# 5. 지각 여부 (함수 작성)
다음과 같이 매개변수와 예시가 주어졌을 때 지각을 하지 않았다면 0을 return하고, 지각하였다면 얼마나 지각하였는지 return하는 함수를 작성하세요.

<hr>

### 매개변수 설명
  * arr = 5분마다 정거장에 버스가 오고, 남은 자리가 표기되어있습니다.<br>
  * start = 문자열로 이루어진 출발시간<br>
  * end = 문자열로 이루어진 도착해야하는 시간<br>
  * duration = 자연수로 이루어진 이동시간(분) <br>
  
<hr>

### return 값 설명
solution 함수는 주어진 arr, start, end, duration을 매개변수로 입력받아 지각을 하지 않았다면 0을 return하고, 지각하였다면 얼마나 지각하였는지 return합니다.

<hr>

### 예시

| arr | start | end | duration | result | explanation |
|---|---|---|---|---|---|
| [0,0,1,0,2,0,3] | "09:00" | "09:30" | 15 | 0 | #2대 놓쳐서 10분+이동15분+출발시간 = 9시25분이므로 지각안함 즉, 0 |
| [0,0,0,0,0,5,8] | "09:00" | "10:00" | 40 | 5 | #5대 놓쳐서 25분+이동40분+출발시간 = 10시05분이므로 지각하고 즉, 5 |
| [0,1,0,2,0,1,3] | "09:00" | "09:10" | 5 | 0 | #1대 놓쳐서 5분+이동5분+출발시간 = 09시10분이므로 지각안함 즉, 0 |

위 예시와 같이 지각을 하지 않았다면 0을 리턴하고, 지각인 경우 얼마나 지각하였는지 리턴하는 함수를 작성하시오.

```python
# import cannot be used

def solution(arr, start, end, duration):
    s_h, s_m = [int(x) for x in start.split(':')]
    e_h, e_m = [int(x) for x in end.split(':')]
    s_m += s_h * 60
    e_m += e_h * 60
    
    for i, v in enumerate(arr):
        if v > 0:
            duration += i * 5
            break
    
    result = s_m + duration - e_m
    answer = 0 if result < 0 else result

    return answer
    
# Below is the code to output the test cases.
print('output_1 = ',solution([0,0,1,0,2,0,3], "09:00", "09:30", 15)) #정답 0
print('output_2 = ',solution([0,0,0,0,0,5,8], "09:00", "10:00", 40)) #정답 5
print('output_3 = ',solution([0,1,0,2,0,1,3], "09:00", "09:10", 5))  #정답 0
arr = [0, 0, 0, 0, 0, 0, 0]
start = "08:00"
end = "09:00"
duration = 30
print('output = ', solution(arr, start, end, duration))  # 예측 결과: 지각으로 5분 이상 (예를 들어, 30+분 지각)
arr = [0, 0, 1, 0, 2, 0, 3]
start = "23:50"
end = "00:30"
duration = 20
print('output = ', solution(arr, start, end, duration))  # 예측 결과: 0 또는 10분 지각
```

# 6. 전구 스위치 (함수 작성)
1번부터 n번까지의 전구 스위치가 있습니다. 스위치를 작동시키면 전구가 꺼져있을땐 켜지고, 켜져있을땐 꺼지도록 설계되어 있습니다.<br>
하지만 특이하게도 2번 스위치를 작동하면 2의 배수인 스위치들이 작동하고, 3번 스위치를 작동하면 3의 배수인 스위치들이 작동한다.<br>

최초에는 모든 전구가 꺼져있다 가정하였을때, 6번 전구인 경우 2번 스위치를 작동시켰을 때 켜지고, 3번 스위치를 작동시켰을 때 꺼졌다.<br>
[0, 0, 0, 0, 0, 0] -> [0, 1, 0, 1, 0, 1] -> [0, 1, 1, 1, 0, 0]

스위치번호가 기록된 배열(switch_list)과 스위치의 개수(n)가 주어졌을 때,<br>
마지막에 켜져있는 전구들의 개수를 구하는 함수를 작성하시오.<br>
[0, 1, 1, 1, 0, 0] -> 켜져있는 전구의 개수 : 3개

<hr>

### 매개변수 설명
  * switch_list는 순서대로 몇번째 스위치들을 작동시켰는지에 대한 이력입니다.
  * switch_list가 [2, 3]인 경우 순차적으로 2번 스위치와 3번 스위치를 작동하였습니다.
  * n은 전구의 개수입니다.

<hr>

### return 값 설명
solution 함수는 주어진 switch_list, n을 매개변수로 입력받아 마지막까지 켜져있는 전구의 개수를 return합니다.
<hr>

### 예시
switch_list = [2, 3], n = 6으로 주어졌을 때, 최초에는 모두 꺼져있는 전구 6개가 있습니다. [0, 0, 0, 0, 0, 0]<br>
2번 스위치를 작동하면, 2의 배수인 2, 4, 6번째 전구가 켜집니다. [0, 1, 0, 1, 0, 1]<br>
3번 스위치를 작동하면, 3의 배수인 3, 6번 전구가 켜지거나 꺼집니다. [0, 1, 1, 1, 0, 0]<br>
마지막까지 켜져있는 전구(2,3,4번째 전구)는 총 3개이므로, 3을 return합니다.<br>


```python
# import cannot be used

def solution(switch_list, n):
    light_arr = [0 for x in range(0, n+1)]
    
    for i in switch_list:
        light_num = i

        while (light_num <= len(light_arr)-1):
            light_arr[light_num] += 1
            light_num += i
    result = [x%2 for x in light_arr]
    answer = sum(result)
    
    return answer

# Below is the code to output the test cases.
print('output_1 = ',solution([2, 3], 6))  # 3
print('output_2 = ',solution([2, 3, 200, 300], 1000))  # 499
print('output_3 = ',solution([2, 8], 1000))  # 375
print('output_4 = ',solution([2, 2], 1000))  # 0
```

# 7. 화단의 꽃 (빈칸 채우기)
n x n 크기 격자 모양 정원에 칸마다 핀 꽃 또는 피지 않은 꽃을 심었습니다. 이 정원의 꽃이 모두 피는 데 며칠이 걸리는지 알고 싶습니다. 핀 꽃은 하루가 지나면 앞, 뒤, 양옆 네 방향에 있는 꽃을 피웁니다. 
정원 크기 n과 현재 정원의 상태를 담은 2차원 리스트 garden이 주어졌을 때, 모든 꽃이 피는데 며칠이 걸리는지 return 하도록 주어진 코드의 빈칸을 적절히 채워 전체 코드를 완성해주세요.

---
### 매개변수 설명
정원 크기 n과 현재 정원 상태를 담은 2차원 리스트 garden이 solution 함수의 매개변수로 주어집니다.
* 정원 크기 n은 1보다 크고 100 보다 작거나 같은 자연수입니다.
* 정원 상태를 담은 2차원 리스트 garden의 원소는 0 또는 1 입니다.
* 이미 핀 꽃은 1로 아직 피지 않은 꽃은 0으로 표현합니다.
* 정원에 최소 꽃 한 개는 피어 있습니다.

---
### return 값 설명
꽃이 모두 피는데 며칠이 걸리는지 return 합니다.

---
### 예시

| n | garden                        	| return |
|---|-----------------------------------|--------|
| 3 | [[0, 0, 0], [0, 1, 0], [0, 0, 0]] | 2  	|
| 2 | [[1, 1], [1, 1]]              	| 0  	|


#### 예시 #1

<br>따라서, 2일이 지나면 정원의 모든 꽃이 핍니다.

#### 예시 #2
첫 날 화단의 상태는 아래와 같습니다. 따라서, 0일이 지나면 정원의 모든 꽃이 핍니다.<br>



```python
# import cannot be used

def func(inp):        
    l = len(inp)
    tmp = [[0]*l for _ in range(l)]
    
    for i in range(l):
        for j in range(l):
            if inp[i][j] == 1:
                if j-1 >= 0:
                    tmp[i][j-1] = 1
                if j+1 < l:
                    tmp[i][j+1] = 1
                if i-1 >= 0:
                    tmp[i-1][j] = 1
                if i+1 < l:
                    tmp[i+1][j] = 1

    sum_lists = [[p+q for p,q in zip(x, y)] for x,y in zip(inp, tmp)]    # sum_lists = [[@@@ in zip(x, y)] for x,y in zip(inp, tmp)]
    merge_lists = [[1 if y==2 else y for y in x] for x in sum_lists]    # merge_lists = [[@@@ in x] for x in tmp_lists]
    
    if all([all(x) for x in merge_lists]) == False:
        return sum_lists, 0
    else:
        return sum_lists, 1

def solution(inp):
    tmp = inp
    result = 0
    
    if all([all(x) for x in inp]):
        return result
    else:
        while True:
            tmp, flag = func(tmp)
            result += 1
            
            if flag == 1:
                return result

# Below is the code to output the test cases.
print('output_1 = ',solution([[0, 0, 0], [0, 1, 0], [0, 0, 0]]))  # 2
print('output_2 = ',solution([[1, 0, 0], [0, 0, 0], [0, 0, 0]]))  # 4
print('output_3 = ',solution([[1, 1, 1], [0, 0, 0], [0, 0, 0]]))  # 2
print('output_4 = ',solution([[1, 1], [1, 1]]))  # 0
```

# 8. 벽 높이에 따른 물의 최소값 (한줄 바꾸기)
아래와 같이 벽이 여러 개 있을 때, 벽 2개를 제외한 나머지 벽을 제거하여 물을 최소한 담으려 합니다. 물은 `두 벽 사이의 거리 x 두 벽 중 낮은 벽의 높이`리터 만큼 담을 수 있으며, 두 벽의 거리는 두 벽의 위치 차이입니다.



모든 벽의 위치와 높이를 담은 2차원 리스트 walls가 매개변수로 주어질 때, 물을 최소 몇 리터나 담을 수 있는지 return 하도록 주어진 코드에서 **한 줄**만 변경해서 모든 입력에 대해 올바르게 동작하도록 수정하세요.

<hr>

### 매개변수 설명

모든 벽의 위치와 높이를 담은 2차원 리스트 walls가 solution 함수의 매개변수로 주어집니다.
* walls는 길이가 3이상 100이하인 2차원 리스트입니다.
* walls의 각 원소는 [벽의 위치, 벽의 높이] 이고, 벽의 위치를 기준으로 오름차순으로 정렬되어 있습니다.
    * 벽의 위치는 1 이상 100 이하입니다.
    * 벽의 높이는 1 이상 100 이하입니다.

<hr>

### return 값 설명

물을 최소 몇 리터나 담을 수 있는지 return 합니다.

<hr>

### 입출력 예

| walls                                 | return |
|------------------------------------------|--------|
| [[1, 4], [2, 6], [3, 5], [5, 3], [6, 2]] | 2     |

아래와 같이 벽을 남겼을 때 물을 최소로 담을 수 있습니다. 이 때 최대 2리터를 담을 수 있습니다.



```python
# import cannot be used

def solution(walls):
    answer = float('inf')
    for i in range(len(walls)):
        for j in range(i+1, len(walls)):
            area =  walls[i][1] * (walls[j][0] - walls[i][0]) if walls[i][1] < walls[j][1] else walls[j][1] * (walls[j][0] - walls[i][0])
            # area =  walls[i][1] * (walls[i][0] - walls[j][0]) if walls[i][1] > walls[j][1] else walls[j][1] * (walls[i][0] - walls[j][0])
            if answer > area:
                answer = area
    return answer

# Below is the code to output the test cases.
#walls = [[1, 4], [2, 6], [3, 5], [5, 3], [6, 2]]
walls = [[3, 1], [4, 6]]
# walls = [[1, 8], [2, 6], [4, 9], [5, 7], [6, 10]]
# walls = [[1, 7], [3, 4], [4, 6], [6, 2], [7, 5]]
# walls = [[1, 5], [3, 3], [5, 6], [7, 4], [8, 7]]
# walls = [[1, 5], [2, 1], [4, 4], [5, 3], [6, 6]]


print('output = ',solution(walls))  # 2
```

## 9. 자아도취 수 (한줄 바꾸기)
어떤 자리 수 k가 주어졌을 때 각 자릿수의 k 제곱의 합이 원래 수가 되는 수를 자아도취 수라고 합니다. 예를 들어 153은 세 자리 자아도취 수입니다.



자연수 k가 매개변수로 주어질 때, k 자리 자아도취 수들을 리스트에 오름차순으로 담아 return 하도록 solution 함수를 작성하려 합니다. 주어진 코드에서 **한 줄**만 변경해서 모든 입력에 대해 올바르게 동작하도록 수정하세요.

<hr>

### 매개변수 설명
k가 solution 함수의 매개변수로 주어집니다.

* k는 3 이상 6 이하인 자연수입니다.

<hr>

### return 값 설명

k 자리 자아도취 수를 오름차순으로 정렬한 뒤 리스트에 담아 return 합니다.

<hr>

### 예시 설명
| k | return               |
|---|----------------------|
| 3 | [153, 370, 371, 407] |

* 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153
* 370 = 3^3 + 7^3 + 0^3 = 27 + 343 + 0 = 370
* 371 = 3^3 + 7^3 + 1^1 = 27 + 343 + 1 = 371
* 407 = 4^3 + 0^3 + 7^3 = 64 + 0 + 343 = 407

```python
# import cannot be used

def solution(k):
	answer = []
	bound = 10**k
    
	for i in range(bound // 10, bound):
		current = i
		calculated = 0
        
		while current != 0:
			calculated += (current%10)**k    # calculated += bound//10
			current //= 10
            
		if calculated == i:
			answer.append(i)
            
	return answer

# Below is the code to output the test cases.
print('output_1 = ',solution(3))  # [153, 370, 371, 407]
print('output_2 = ',solution(4))  # [1634, 8208, 9474]
```

# 10. 비숍으로부터 도망쳐 (빈칸 채우기)
체스에서 비숍(Bishop)은 아래 그림과 같이 대각선 방향으로 몇 칸이든 한 번에 이동할 수 있습니다. 만약, 한 번에 이동 가능한 칸에 체스 말이 놓여있다면 그 체스 말을 잡을 수 있습니다.



8 x 8 크기의 체스판 위에 여러 개의 비숍(Bishop)이 놓여있습니다. 이때, 비숍(Bishop)들에게 _**한 번에**_ 잡히지 않도록 새로운 말을 놓을 수 있는 빈칸의 개수를 구하려고 합니다.

위 그림에서 원이 그려진 칸은 비숍에게 한 번에 잡히는 칸들이며, 따라서 체스 말을 놓을 수 있는 빈칸 개수는 50개입니다.

8 x 8 체스판에 놓인 비숍의 위치 bishops가 매개변수로 주어질 때, 비숍에게 한 번에 잡히지 않도록 새로운 체스 말을 놓을 수 있는 빈칸 개수를 return 하도록 solution 함수의 빈칸(@@@)을 적절히 채워 전체 코드를 완성해주세요.

<hr>

### 매개변수 설명
체스판에 놓인 비숍의 위치 bishops가 solution 함수의 매개변수로 주어집니다.
* bishops는 비숍의 위치가 문자열 형태로 들어있는 리스트입니다.
* bishops의 길이는 1 이상 64 이하입니다.
* 비숍이 놓인 위치는 알파벳 대문자와 숫자로 표기합니다.
  * 알파벳 대문자는 가로 방향, 숫자는 세로 방향 좌표를 나타냅니다.
  * 예를 들어 위 그림에서 비숍이 있는 칸은 "D5"라고 표현합니다.
* 한 칸에 여러 비숍이 놓이거나, 잘못된 위치가 주어지는 경우는 없습니다.

<hr>

### return 값 설명
비숍에게 한 번에 잡히지 않도록 새로운 체스 말을 놓을 수 있는 빈칸의 개수를 return 해주세요.

<hr>

### 예시

| bishops            | return |
|--------------------|--------|
| ["D5"]             | 50     |
| ["D5", "E8", "G2"] | 42     |

예시 #1
문제에 나온 예시와 같습니다.

예시 #2



그림과 같이 원이 그려진 칸은 비숍에게 한 번에 잡히는 칸들이며, 따라서 체스 말을 놓을 수 있는 빈칸 개수는 42개입니다.

```python
# import cannot be used

def solution(bishops):
    answer = 0
    result = []
    bishop_can = [(-1, 1), (-1, -1), (1, -1), (1, 1)]
    for chess_p in bishops:
        x_pos = int(ord(chess_p[0])-ord('A')+1)    # x_pos = @@@
        y_pos = int(chess_p[1])
        result.append((x_pos,y_pos))
        for way in bishop_can:
            x, y = x_pos, y_pos
            while 0 < x < 9 and 0 < y < 9:
                x += way[0]
                y += way[1]
                if 0 < x < 9 and 0 < y < 9:
                    result.append((x, y))    # result.@@@
    result = set(result)     # result = @@@
    answer = 64 - len(result)
    return answer

# Below is the code to output the test cases.
print('output_1 = ',solution(["D5"]))  # 50
print('output_2 = ',solution(["D5", "E8", "G2"]))  # 42
```
