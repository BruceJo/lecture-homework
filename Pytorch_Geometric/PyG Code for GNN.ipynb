```python
from torch_geometric.nn.conv import MessagePassing
```

```python
class MLP(nn.Module): #FNN을 생성하는 module입니다. 
    def __init__(self, input_dim, hid_dims, out_dim):
        super(MLP, self).__init__()

        self.mlp = nn.Sequential()
        dims = [input_dim] + hid_dims + [out_dim]
        for i in range(len(dims)-1):
            self.mlp.add_module('lay_{}'.format(i),nn.Linear(in_features=dims[i], out_features=dims[i+1]))
            if i+2 < len(dims):
                self.mlp.add_module('act_{}'.format(i), nn.Tanh())
    def reset_parameters(self):
        for i, l in enumerate(self.mlp):
            if type(l) == nn.Linear:
                nn.init.xavier_normal_(l.weight)

    def forward(self, x):
        return self.mlp(x)
```

```python
class GatedGraphConv(MessagePassing):
    
    def __init__(self, out_channels, num_layers, aggr = 'add',
                 bias = True, **kwargs):
        super(GatedGraphConv, self).__init__(aggr=aggr, **kwargs)

        self.out_channels = out_channels #
        self.num_layers = num_layers

        self.weight = Param(Tensor(num_layers, out_channels, out_channels))
        self.rnn = torch.nn.GRUCell(out_channels, out_channels, bias=bias)
		#GNN의 rnn부분을 GRU를 사용하여 정의했습니다. 
        self.reset_parameters()

    def reset_parameters(self):
        uniform(self.out_channels, self.weight)
        self.rnn.reset_parameters()

    def forward(self, data):
        data= data.to(device) #해줘야 cuda로 보낼 수 있음         
        x = data.x
        edge_index = data.edge_index #edge 부분
        edge_weight = data.edge_attr #edge_weight부분 
        if x.size(-1) > self.out_channels: #예외 처리 
            raise ValueError('The number of input channels is not allowed to '
                             'be larger than the number of output channels')

        if x.size(-1) < self.out_channels:
            zero = x.new_zeros(x.size(0), self.out_channels - x.size(-1))
            x = torch.cat([x, zero], dim=1)

        for i in range(self.num_layers):
            m = torch.matmul(x, self.weight[i])
            m = self.propagate(edge_index, x=m, edge_weight=edge_weight,
                               size=None)
            x = self.rnn(m, x)

        return x

    def message(self, x_j, edge_weight):
        return x_j if edge_weight is None else edge_weight.view(-1, 1) * x_j

    def message_and_aggregate(self, adj_t, x):
        return matmul(adj_t, x, reduce=self.aggr)

    def __repr__(self):
        return '{}({}, num_layers={})'.format(self.__class__.__name__,
                                              self.out_channels,
                                              self.num_layers)
```

```python
from torch_geometric.nn.conv import GatedGraphConv
from torch_geometric.nn.models import MLP
```

```python
class GGNN(torch.nn.Module):
    def __init__(self):
        super(GGNN, self).__init__()
        
        self.ggnn = GatedGraphConv(1433, 3).to(device)
        self.mlp = MLP(1433, [32,32,32], dataset.num_classes).to(device)
        
    def forward(self):
        x = self.ggnn(data)
        x = self.mlp(x)
        return F.softmax(x, dim=-1)

    def reset_parameters(self):
        self.ggnn.reset_parameters()
        self.mlp.reset_parameters()
```

```python
model = GGNN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss = nn.CrossEntropyLoss()
```

```python
# 학습 루프 정의 
def train():
    model.train() #학습
    optimizer.zero_grad() # 이전 epoch에서 역전파 시 계산된 기울기를 0으로 초기화
    data.train_mask = data.train_mask.to(device) # train data를 정의 및 cuda 사용
    loss(model()[data.train_mask], data.y[data.train_mask]).backward() 
    # 파라미터들의 에러에 대한 변화도를 계산하여 누적함
    optimizer.step() # optimizer에게 파라미터 수정 지시


def test():
    model.eval() # 모델을 평가 모드로 전환 
    logits, accs = model(), []
    for _, mask in data('train_mask', 'val_mask', 'test_mask'):
        pred = logits[mask].max(1)[1]
        acc = pred.eq(data.y[mask]).sum().item() / mask.sum().item()
        accs.append(acc)
    return accs

model.reset_parameters() # 훈련 재개 시, 이전 학습에 사용된 파라미터를 초기화

# 25번의 epoch 동안 모델을 학습 
for epoch in range(25):
    train()
    accs = test()
    train_acc = accs[0]
    val_acc = accs[1]
    test_acc = accs[2]
    print('Epoch: {:02d}, Train Acc: {:.5f}, '
          'Val Acc: {:.5f}, Test Acc: {:.5f}'.format(epoch+1, train_acc, val_acc, test_acc))
```

```python
model = GGNN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss = nn.CrossEntropyLoss()
```

```python
# 학습 루프 정의 
def train():
    model.train() #학습
    optimizer.zero_grad() # 이전 epoch에서 역전파 시 계산된 변화도를 0으로 초기화
    data.train_mask = data.train_mask.to(device) # train data를 정의 및 cuda 사용
    loss(model()[data.train_mask], data.y[data.train_mask]).backward() 
    # train의 y를 주고, 가중치의 에러에 대한 변화도를 계산하여 누적함
    optimizer.step() # optimizer에게 파라미터 수정 지시


def test():
    model.eval() # 모델을 평가 모드로 전환
    accs = []
    masks = [data.train_mask, data.val_mask, data.test_mask]  # 마스크 정의
    
    for mask in masks:  # 마스크 순회
        pred = model()[mask].max(1)[1]
        '''
        train()함수가 수행된 후 model()에 pred는 이미 계산되어 있음
        여기선, 슬라이싱하여 pred를 가져옴
        max(1)은 가장 높은 확률로 예측된 클래스
        [0]은 softmax의 출력값, [1]은 예측된 클래스
        '''
        acc = pred.eq(data.y[mask]).sum().item() / mask.sum().item()
        accs.append(acc)
    return accs
```

```python
model.reset_parameters() # 훈련 재개 시, 이전 학습에 사용된 파라미터를 초기화

# 25번의 epoch 동안 모델을 학습 
for epoch in range(25):
    train()
    accs = test()
    print('Epoch: {:02d}, Train Acc: {:.5f}, '
          'Val Acc: {:.5f}, Test Acc: {:.5f}'.format(epoch+1, accs[0], accs[1], accs[2]))
```
