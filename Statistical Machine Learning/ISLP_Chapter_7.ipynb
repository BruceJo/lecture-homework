### <font color=green>Chapter 7.P1.</font>
<font color=green>
It was mentioned in this chapter that a cubic regression spline with

one knot at , can be obtained using a basis of the form $x, x2, x3,(x − ξ)^3_+$

where $(x − ξ)^3_+ = (x − ξ)^3$ if $x > ξ$ and equals 0 otherwise.

We will now show that a function of the form

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β4(x − ξ)^3_+
$$

is indeed a cubic regression spline, regardless of the values of $β_0, β_1, β_2,
β_3, β_4$.
</font>

### <font color=green>7.P1.(a) </font>
<font color=green>
Find a cubic polynomial

$$
f_1(x) = a_1 + b_1x + c_1x^2 + d_1x^3
$$

such that $f(x) = f_1(x)$ for all $x ≤ ξ$. Express $a_1, b_1, c_1, d_1$ in

terms of $β_0, β_1, β_2, β_3, β_4$.
</font>

### answer 7.P1.(a)

주어진 함수는

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β4(x − ξ)^3_+
$$

여기서 $(x − ξ)^3_+$는 truncated power basis function으로,

$$
(x − ξ)^3_+ =
\begin{cases} 
(x − ξ)^3, & \text{if } x>ξ \\\\
0, & \text{if } x≤ξ
\end{cases}
$$

따라서 $x≤ξ$일 때는 $(x − ξ)^3_+ =0$이므로:

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3
$$

위 식은 3차 다항식이며, 다음과 같이 비교가 가능:

$a_1=β_0$

$b_1=β_1$

$c_1=β_2$

$d_1=β_3$

### <font color=green>7.P1.(b) </font>
<font color=green>
Find a cubic polynomial

$$
f_2(x) = a_2 + b_2x + c_2x^2 + d_2x^3
$$

such that $f(x) = f_2(x)$ for all $x > \xi$. Express $a_2, b_2, c_2, d_2$ in terms of $β_0, β_1, β_2, β_3, β_4$.

We have now established that $f(x)$ is a piecewise polynomial.
</font>

### answer 7.P1.(b)

주어진 함수는

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β4(x − ξ)^3_+
$$

여기서 $x > \xi$일 때는 $(x - \xi)^3_+ = (x - \xi)^3$이므로:

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β4(x − ξ)^3
$$

$(x - \xi)^3$을 전개하면 다음과 같습니다.

$$
(x − ξ)^3 = x^3 − 3ξx^2 + 3ξ^2x − ξ^3
$$

이를 대입하여 정리하면:

$$
f(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β4(x^3 − 3ξx^2 + 3ξ^2x − ξ^3)
$$
$$
=(β_0 - β_4\xi^3) + (β_1 + 3β_4\xi^2)x + (β_2 - 3β_4\xi)x^2 + (β_3 + β_4)x^3
$$

따라서, 다음과 같이 계수를 정리할 수 있습니다:

$a_2 = \beta_0 - \beta_4 \xi^3$

$b_2 = \beta_1 + 3\beta_4 \xi^2$

$c_2 = \beta_2 - 3\beta_4 \xi$

$d_2 = \beta_3 + \beta_4$

### <font color=green>7.P1.(c) </font>
<font color=green>Show that $f_1(\xi) = f_2(\xi)$. That is, $f(x)$ is continuous at $\xi$.</font>

### answer 7.P1.(c)

먼저, $f_1(x)$는 $x \le \xi$일 때의 표현이므로:

$$
f_1(x) = β_0 + β_1x + β_2x^2 + β_3x^3
$$

따라서,

$$
f_1(\xi) = β_0 + β_1\xi + β_2\xi^2 + β_3\xi^3
$$

다음으로, $f_2(x)$는 $x > \xi$일 때의 표현이며, 앞에서 구한 바에 따르면:

$$
f_2(x) = (β_0 - β_4\xi^3) + (β_1 + 3β_4\xi^2)x + (β_2 - 3β_4\xi)x^2 + (β_3 + β_4)x^3
$$

여기에서 $x = \xi$를 대입하면:
$$
f_2(\xi) = (β_0 - β_4\xi^3) + (β_1 + 3β_4\xi^2)\xi + (β_2 - 3β_4\xi)\xi^2 + (β_3 + β_4)\xi^3
$$
$$
= β_0 - β_4\xi^3 + β_1\xi + 3β_4\xi^3 + β_2\xi^2 - 3β_4\xi^3 + β_3\xi^3 + β_4\xi^3
$$

이제 항을 정리하면:
$$
f_2(\xi) = β_0 + β_1\xi + β_2\xi^2 + β_3\xi^3
$$

따라서 함수 $f(x)$는 $x=ξ$에서 연속(continuous).

$$
f_1(\xi) = f_2(\xi)
$$

### <font color=green>7.P1.(d) </font>
<font color=green>Show that $f_1'(ξ) = f_2'(ξ)$. That is, $f'(x)$ is continuous at $ξ$.</font>

### answer 7.P1.(d)

다음 $f_1(x)$에 대하여
$$f_1(x) = β_0 + β_1x + β_2x^2 + β_3x^3$$

미분하면:
$$f_1'(x) = β_1 + 2β_2x + 3β_3x^2$$

$x=\xi$에 따라:
$$f_1'(\xi) = β_1 + 2β_2\xi + 3β_3\xi^2$$

$f_2(x)$에 대하여
$$f_2(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β_4(x-\xi)^3$$

미분하면:
$$f_2'(x) = β_1 + 2β_2x + 3β_3x^2 + 3β_4(x-\xi)^2$$

$x=\xi$에 따라:
$$f_2'(\xi) = β_1 + 2β_2\xi + 3β_3\xi^2 + 3β_4(\xi-\xi)^2$$

$x = \xi$에서 마지막 항은 $(x - \xi)^2 = 0$이므로:
$$f_2'(\xi) = β_1 + 2β_2\xi + 3β_3\xi^2$$

따라서 함수 $f'(x)$는 $x=ξ$에서 연속(continuous).:
$$f_1'(ξ) = f_2'(ξ)$$

### <font color=green>7.P1.(e) </font>
<font color=green>Show that $f_1''(ξ) = f_2''(ξ)$. That is, $f''(x)$ is continuous at $ξ$.</font>

### answer 7.P1.(e)

다음 $f_1(x)$에 대하여
$$f_1(x) = β_0 + β_1x + β_2x^2 + β_3x^3$$

2차 미분:
$$f_1''(x) = 2β_2 + 6β_3x$$

$x=\xi$에 따라:
$$f_1''(\xi) = 2β_2 + 6β_3\xi$$

$f_2(x)$에 대하여
$$f_2(x) = β_0 + β_1x + β_2x^2 + β_3x^3 + β_4(x-\xi)^3$$

2차 미분:
$$f_2''(x) = 2β_2 + 6β_3x + 6β_4(x-\xi)$$

$x=\xi$에 따라:
$$f_2''(\xi) = 2β_2 + 6β_3\xi + 6β_4(\xi-\xi)$$
$$= 2β_2 + 6β_3\xi$$
$$f_1''(\xi) = f_2''(\xi)$$

따라서 함수 $f''(x)$는 $x=ξ$에서 연속(continuous).

즉, 이 함수는 cubic regression spline

### <font color=green>Chapter 7.P2.</font>
<font color=green>
Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of $n$ points using the following formula:

$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int [g^{(m)}(x)]^2 \, dx \right)
$$
where $g^{(m)}$ represents the $m$th derivative of $g$ (and $g^{(0)} = g$). Provide
example sketches of $\hat{g}$ in each of the following scenarios.
</font>

```python
# 해당 문제에서는 아래 데이터를 사용하여 g_hat의 변화를 관찰
# 사용 예시는 Y=g(X)+e
# g(X) = 1/X+0.2 * sin(5(X+0.2))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.uniform(0, 1, 100)
epsilon = np.random.normal(0, 1, 100)
y = np.sin(5 * (x + 0.2)) / (x + 0.2) + epsilon

def generating_function(x_vals):
    return np.sin(5 * (x_vals + 0.2)) / (x_vals + 0.2)

df = pd.DataFrame({'x': x, 'y': y})

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>7.P2.(a) </font>
<font color=green>$\lambda = \infty, m = 0$.</font>

### answer 7.P2.(a)

$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int [g(x)]^2 \, dx \right)
$$

$\lambda$가 증가함에 따라 위 식에서 패널티항이 중요해짐

이 경우, $g(x)$가 0이 아닌 어떤 값을 가지면 패널티가 무한대가 되므로, $g(x)$는 0이 되어야함

이는 데이터와 무관하게, 오직 페널티 항을 최소화

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)
ax.axhline(y=0, color='orange')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>7.P2.(b)</font>
<font color=green>$\lambda = \infty, m = 1$.</font>

### answer 7.P2.(b)
$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int [g'(x)]^2 \, dx \right)
$$

$\lambda=\infty$일 때 $g'(x) = 0$이어야 하므로, 즉 상수 함수가 되어야 함

이 값들은 $y_i$들의 평균

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)
ax.axhline(y=np.mean(y), color='orange')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>7.P2.(c) </font>
<font color=green>$\lambda = \infty, m = 2$.</font>

### answer 7.P2.(c)
$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int [g''(x)]^2 \, dx \right)
$$

$\lambda=\infty$일 때 $g''(x) = 0$이어야 하므로, $\hat{g}=ax+b$

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)

y_pred = np.polyval(np.polyfit(x, y, deg=1), x_grid)
ax.plot(x_grid, y_pred, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>7.P2.(d)</font>
<font color=green>$\lambda = \infty, m = 3$.</font>

### answer 7.P2.(d)

$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int \left[ g^{(3)}(x) \right]^2 \, dx \right)
$$

$\lambda=\infty$일 때 $g^{(3)}(x) = 0$이어야 하므로, $\hat{g}=ax^2+bx+c$

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)

y_pred = np.polyval(np.polyfit(x, y, deg=2), x_grid)
ax.plot(x_grid, y_pred, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>7.P2.(e) </font>
<font color=green>$\lambda = 0, m = 3$.</font>

### answer 7.P2.(e)

$$
\hat{g} = \arg\min_{g} \left( \sum_{i=1}^{n} (y_i - g(x_i))^2 + \lambda \int \left[ g^{(3)}(x) \right]^2 \, dx \right)
$$

$\lambda=0$에서 패널티 항이 완전히 제거되어 오직 데이터의 적합만 고려됨

$$
\hat{g} = \arg\min_{g} \sum_{i=1}^{n} (y_i - g(x_i))^2
$$

이 spline은 모든 $x_i$에서 연속적인 함수가 되어 매우 요동치는 함수가 될 수 있음

```python
from scipy.interpolate import UnivariateSpline

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['x'], df['y'], alpha=0.5)
x_grid = np.linspace(0, 1, 500)
y_true = generating_function(x_grid)
ax.plot(x_grid, y_true)

sort_idx = np.argsort(x)
x_sorted = x[sort_idx]
y_sorted = y[sort_idx]

spline = UnivariateSpline(x_sorted, y_sorted, s=0)
x_grid = np.linspace(x_sorted.min(), x_sorted.max(), 500)
y_spline = spline(x_grid)

ax.plot(x_grid, y_spline, color='orange')
ax.set_ylim(-3, 5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
plt.show()
```

### <font color=green>Chapter 7.P9.</font>
<font color=green>
This question uses the variables dis (the weighted mean of distances
to five Boston employment centers) and nox (nitrogen oxides concentration
in parts per 10 million) from the Boston data. We will treat
dis as the predictor and nox as the response.
</font>

### <font color=green>7.P9.(a) </font>
<font color=green>
Use the poly() function from the ISLP.models module to fit a
cubic polynomial regression to predict nox using dis. Report the
regression output, and plot the resulting data and polynomial
fits.
</font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ISLP import load_data, models
from ISLP.models import poly, ModelSpec, summarize, sklearn_sm, bs
import statsmodels.api as sm
from sklearn.model_selection import cross_validate, KFold
```

```python
# answer 7.P9.(a)

boston = load_data('Boston')
dis=boston['dis']
nox=boston['nox']

model = ModelSpec([poly('dis', degree=3)]).fit(boston)
X = model.transform(boston)
result = sm.OLS(nox, X).fit()
summarize(result)
```

```python
x_grid = np.linspace(dis.min(), dis.max(), 300)
x_grid_df = pd.DataFrame(x_grid, columns=['dis'])
origin_x_grid_df = pd.DataFrame(dis, columns=['dis'])

x_grid_poly = model.transform(x_grid_df)
y_preds = result.get_prediction(x_grid_poly)

bands = y_preds.conf_int(alpha=0.05)

plt.figure(figsize=(8, 5))
plt.scatter(dis, nox, alpha=0.4, facecolor='gray')

for val, ls in zip([y_preds.predicted_mean,bands[:,0],bands[:,1]],['b','r--','r--']):
    plt.plot(x_grid_df.values, val, ls, linewidth=3)

plt.xlabel('dis')
plt.ylabel('nox')
plt.grid(True)
plt.show()
```

적합성이 좋아 보이지만, dis의 끝 부분에서는 한계가 보임

### <font color=green>7.P9.(b) </font>
<font color=green>
Plot the polynomial fits for a range of different polynomial
degrees (say, from 1 to 10), and report the associated residual
sum of squares.</font>

```python
# answer 7.P9.(b)

x_grid = np.linspace(dis.min(), dis.max(), 300)
x_grid_df = pd.DataFrame(x_grid, columns=['dis'])
rss_list = []

fig, axes = plt.subplots(5, 2, figsize=(16, 24))
axes = axes.flatten()

for degree in range(1, 11):
    model = ModelSpec([poly('dis', degree=degree)]).fit(boston)
    X = model.transform(boston)
    result = sm.OLS(nox, X).fit()
    
    x_grid_poly = model.transform(x_grid_df)
    y_pred = result.predict(x_grid_poly)
    
    rss = result.ssr
    rss_list.append(rss)
    
    ax = axes[degree - 1]
    ax.scatter(dis, nox, alpha=0.4, color='gray')
    ax.plot(x_grid, y_pred, linewidth=2)
    ax.set_title(f"Degree {degree}, RSS = {rss:.4f}")
    ax.set_xlabel("dis")
    ax.set_ylabel("nox")
    ax.grid(True)

plt.show()
```

3차와 4차가 시각적으로 가장 타당해 보임

### <font color=green>7.P9.(c) </font>
<font color=green>
Perform cross-validation or another approach to select the optimal
degree for the polynomial, and explain your results.</font>

```python
# answer 7.P9.(c)

cv_errors = []
model = sklearn_sm(sm.OLS)
cv = KFold(n_splits=5, shuffle=True, random_state=0)

for degree in range(1, 11):
    X_poly = np.power.outer(np.array(dis), np.arange(degree + 1))
    scores = cross_validate(model, X_poly, nox, cv=cv)
    cv_errors.append(np.mean(scores['test_score']))

for d, err in enumerate(cv_errors):
    print(f"Degree {d+1:02} CV MSE = {err:.4f}")

print(f"\nThe selected model: {cv_errors.index(min(cv_errors))+1}")

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), cv_errors, marker='o')
plt.xlabel('Polynomial Degree')
plt.ylabel('CV MSE')
plt.grid(True)
plt.show()
```

교차검증을 통해 선택된 최적의 다항식 모델 차수는 3차였으며, 이는 underfitting과 overfitting 사이의 균형을 잘 잡는 모델로 판단됨

특히 차수가 증가함에 따라 교차검증 오차가 일시적으로 감소하다가 다시 증가하는 스파이크 현상이 관찰되었으며, 

이는 고차 다항식 모델이 데이터에 과도하게 적합(overfit)되며 발생한 것으로 해석됨

이는 훈련 데이터가 해당 영역에서 희소하기 때문으로 보임

### <font color=green>7.P9.(d) </font>
<font color=green>
Use the bs() function from the ISLP.models module to fit a regression
spline to predict nox using dis. Report the output for
the fit using four degrees of freedom. How did you choose the
knots? Plot the resulting fit.
</font>

```python
# answer 7.P9.(d)

model = ModelSpec([bs('dis', df=4)]).fit(boston)
X = model.transform(boston)
x_grid_df = pd.DataFrame({'dis':np.linspace(dis.min(), dis.max(), 100)})
x_grid_poly = model.transform(x_grid_df)
model = sm.OLS(nox, X).fit()
y_preds = model.get_prediction(x_grid_poly)

bands = y_preds.conf_int(alpha=0.05)

fig, ax = plt.subplots(figsize=(8, 5))
plt.scatter(dis, nox, alpha=0.4, facecolor='gray')

for val, ls in zip([y_preds.predicted_mean,bands[:,0],bands[:,1]],['b','r--','r--']):
    ax.plot(x_grid_df.values, val, ls, linewidth=3)

plt.xlabel('dis')
plt.ylabel('nox')
plt.grid(True)
plt.show()
```

지정된 열에 대해 B-spline basis를 생성하고, 이를 회귀에 사용할 수 있는 특성(Feature) 객체로 반환

knots를 명시적으로 주지 않으면, 내부적으로 데이터의 분위수(quantile) 기준으로 균등하게 분할하여 knot 위치를 설정

따라서, knots는 bs() 함수에서 자동으로 선택되었으며, 이는 주어진 자유도 에 맞추어 입력 변수(dis)의 분위수를 기준으로 균등하게 설정됨

자유도 4의 경우, 적절한 수의 내부 매듭을 선택하여 총 4개의 basis 함수가 만들어지도록 구성됨

### <font color=green>7.P9.(e) </font>
<font color=green>
Now fit a regression spline for a range of degrees of freedom, and
plot the resulting fits and report the resulting RSS. Describe the
results obtained.</font>

```python
# answer 7.P9.(e)

x_grid_df = pd.DataFrame({'dis': np.linspace(dis.min(), dis.max(), 300)})
rss_list = []

fig, axes = plt.subplots(4, 2, figsize=(16, 24))
axes = axes.flatten()

for degree in range(3, 11):
    ax = axes[degree-3]
    model = ModelSpec([bs('dis', df=degree)]).fit(boston)
    X = model.transform(boston)
    result = sm.OLS(nox, X).fit()
    
    X_pred = model.transform(x_grid_df)
    y_pred = result.get_prediction(X_pred)
    y_mean = y_pred.predicted_mean
    
    rss = np.sum((nox - result.fittedvalues) ** 2)
    rss_list.append((degree, rss))
    
    ax.scatter(dis, nox, alpha=0.4, color='gray')
    ax.plot(x_grid_df['dis'], y_mean, linewidth=2)
    ax.set_title(f'df = {degree}, RSS = {rss:.4f}')
    ax.set_xlabel('dis')
    ax.set_ylabel('nox')
    ax.grid(True)

plt.show()

for degree, rss in rss_list:
    print(f'df = {degree:02} RSS = {rss:.4f}')
```

### <font color=green>7.P9.(f) </font>

Perform cross-validation or another approach in order to select
the best degrees of freedom for a regression spline on this data.
Describe your results.</font>

```python
# answer 7.P9.(f)

cv_errors = []
model = sklearn_sm(sm.OLS)
cv = KFold(n_splits=5, shuffle=True, random_state=0)

for degree in range(3, 14):
    spec = ModelSpec([bs('dis', df=degree)]).fit(boston)
    X = spec.transform(boston)
    result = cross_validate(model, X, nox.values, cv=cv)
    cv_errors.append(np.mean(result['test_score']))
```

```python
for d, err in enumerate(cv_errors):
    print(f"Degree {d+3:02} CV MSE = {err:.6f}")

print(f"\nThe selected model: {cv_errors.index(min(cv_errors))+1}")

plt.figure(figsize=(8, 5))
plt.plot(range(3, 14), cv_errors, marker='o')
plt.xlabel('Degrees of Freedom(df)')
plt.ylabel('CV MSE')
plt.grid(True)
plt.show()
```

```python
# answer 7.P9.(g)

dis_centered = dis - 4
dis_above_4 = np.where(dis > 4, dis_centered, 0)

X_piecewise = sm.add_constant(np.column_stack((dis, dis_above_4)))

model_pw = sm.OLS(nox, X_piecewise).fit()
rss_pw = np.sum((model_pw.fittedvalues - nox) ** 2)
print(model_pw.summary())
print(f"Piecewise Linear RSS = {rss_pw:.4f}")
```

```python
x_vals = np.linspace(dis.min(), dis.max(), 300)
x_above_4 = np.where(x_vals > 4, x_vals - 4, 0)
X_pred = sm.add_constant(np.column_stack((x_vals, x_above_4)))
y_pred = model_pw.predict(X_pred)

plt.figure(figsize=(8, 6))
plt.scatter(dis, nox, alpha=0.4, color='gray')
plt.plot(x_grid_df['dis'], y_mean, linewidth=2, color='blue', label=f'final model in (f) RSS:{rss:.4f}')
plt.plot(x_vals, y_pred, linewidth=2, color='red', label=f'Piecewise Linear RSS:{rss_pw:.4f}')
plt.xlabel('dis')
plt.ylabel('nox')
plt.legend()
plt.grid(True)
plt.show()
```
