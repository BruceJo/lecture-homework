(a) The model is defined as:

$$

\begin{cases} 
\alpha_0 + \alpha_1 x + \epsilon, & \text{if } x \leq x_0 \\\\
\beta_0 + \beta_1 x + \epsilon, & \text{if } x > x_0 
\end{cases}
$$

If $x_0$ is known, the model can be rewritten as a linear model with a suitable choice of explanatory variables. 

Show that it can be expressed in the form:

$$
y = \alpha_0 + \alpha_1 Z_1 + \beta_1 Z_2 + \gamma_1 Z_3 + \epsilon
$$

Identify $Z_1$, $Z_2$, $Z_3$, and $\gamma_1$.

Note: Usually the change point $x_0$ is unknown, making the piecewise linear model
nonlinear.

=> 모델은 두 구간으로 나뉘며, $x > x_0$ 일 때만 다른 계수와 절편을 갖는다. 

이를 하나의 식으로 표현하기 위해 $x > x_0$인 경우에만 영향을 주는 항들을 추가한다. 

이때 $Z_2 = (x - x_0), Z_3 = 1$(단, $x > x_0$일 때만)로 정의하면, 전체 모델을 선형 형태로 표현할 수 있다.

여기서 $Z_1$, $Z_2$, $Z_3$, and $\gamma_1$는 다음과 같이 정의될 수 있다 :

$Z_1 = x$    (모든 데이터에 대해 동일)

$Z_2 = (x-x_0)*1(x>x_0)$    (단, $x > x_0$일 때만, 그렇지 않으면 0)

$Z_3 = 1(x>x_0)$    (단, $x > x_0$일 때만, 그렇지 않으면 0)

이를 바탕으로 $x > x_0$일 때 $Z_2 = (x - x_0), Z_3 = 1$ 이므로:

$$
y = \alpha_0 + \alpha_1 x + \beta_1 (x-x_0) + \gamma_1 + \epsilon
$$
$$
= (\alpha_0 + \gamma_1 - \beta_1 x_0) + (\alpha_1 + \beta_1)x + \epsilon
$$

위 식이 다음과 같아야함 :
$$y = \beta_0 + \beta_1 x + \epsilon$$

따라서 두 식이 같아지려면 다음 조건들이 필요 :
$$
\begin{cases} 
\beta_0 = \alpha_0 + \gamma_1 - \beta_1 x_0 \\\\
\beta_1 = \alpha_1 + \beta_1
\end{cases}
$$

첫번째 식만 정리하면 :
$$
\gamma_1 = \beta_0 - \alpha_0 + \beta_1 x_0
$$

(c) Set a random seed, and then compute the LOOCV errors that result from fitting the following four models using least squares:

i. $Y = β0 + β1X + ϵ$

ii. $Y = β0 + β1X + β2X2 + ϵ$

iii. $Y = β0 + β1X + β2X2 + β3X3 + ϵ$

iv. $Y = β0 + β1X + β2X2 + β3X3 + β4X4 + ϵ$

Note you may find it helpful to use the data.frame() function to create a single data set containing both $X$ and $Y$.

````python
import numpy as np
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = x - 2 * x**2 + rng.normal(size=100)
x = x.reshape(-1, 1)

def loocv_err(degree, x, y):
    loo = LeaveOneOut()
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(x)
    errors = []

    for train_index, test_index in loo.split(X_poly):
        X_train, X_test = X_poly[train_index], X_poly[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model = LinearRegression()
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        errors.append(mean_squared_error(y_test, y_pred))

    return np.mean(errors)

for degree in range(1, 5):
    error = loocv_err(degree, x, y)
    print(f"Degree {degree}: LOOCV Error = {error:.4f}")
````

(d) Repeat (c) using another random seed, and report your results.

````python
rng = np.random.default_rng(123)
x = rng.normal(size=100)
y = x - 2 * x**2 + rng.normal(size=100)
x = x.reshape(-1, 1)

for degree in range(1, 5):
    error = loocv_err(degree, x, y)
    print(f"Degree {degree}: LOOCV Error = {error:.4f}")
````

이는 예상된 결과로, 데이터는 다음과 같은 2차 함수 모델에 따라 생성되었기 때문 :

$$y=x-2x^2+\epsilon$$

따라서 2차 다항 회귀 모델이 가장 잘 맞는 함수형태이며,

1차 모델은 과소적합(underfitting), 3차 이상 모델은 과적합(overfitting) 경향이 나타날 수 있음

실제로도 3차, 4차 모델은 오차가 약간 증가

(f) Comment on the statistical significance of the coefficient estimates that results from fitting each of the models in (c) using least squares. 

Do these results agree with the conclusions drawn based on the cross-validation results?

````python
import statsmodels.api as sm

rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = x - 2 * x**2 + rng.normal(size=100)

for degree in [1, 2, 3, 4]:
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly.fit_transform(x.reshape(-1, 1))
    X_poly = sm.add_constant(X_poly)

    model = sm.OLS(y, X_poly).fit()
    print(f"Degree {degree}:")
    print("Coefficients:", np.round(model.params, 3))
    print("P-values:", np.round(model.pvalues, 3))
    print("T-values:", np.round(model.tvalues, 3), "\n")

````

=> 각 다항 회귀 모델에 대해 최소제곱법을 적용하고, 회귀 계수들의 통계적 유의성을 평가

> 1차 모델

>>
>> 그러나 실제 데이터 생성 모델이 2차 함수이므로, 이는 잘못된 단순 적합의 가능성이 있음

> 2차 모델
>> 상수항(p = 0.543)은 유의하지 않으나, $x, x^2& 항은 p-value = 0.000으로 매우 유의미

> 3차 모델
>> $x^3$항의 p-value = 0.287 은 유의하지 않음
>> 
>> 나머지 항은 유의미하지만, 불필요한 항이 포함되어 과적합 가능성 존재

> 4차 모델
>> $x^3$ 항은 여전히 유의하지 않으며 (p = 0.642), $x^4$ 항은 p = 0.023로 다소 유의
>>
>> 그러나 전체적으로 불필요한 고차항이 추가되어 모델 복잡도가 증가함
